```python
#1a
def probabilitateAcZiNastere(n, repetari):
    total = 0
    for j in range(repetari): 
        zile = [0 for _ in range(367)]
        found = False
        for i in range(n):
            zi = randint(1,366)
            zile[zi] += 1
            if zile[zi] > 1:
                found = True
                break
        if found:
            total += 1 
    return total / repetari   
```

```python
print(probabilitateAcZiNastere(23,4000))
```

```python
#2a
def punctePatrat(n):
    axis('square')
    nrp = 0
    axis((0, 1, 0, 1))
    for _ in range(n):
        F=[random(),random()]
        distance_squared = (F[0] - 0.5) ** 2 + (F[1] - 0.5) ** 2
        if distance_squared <= 0.5 ** 2: 
            plot(F[0],F[1],'ro')
            nrp += 1
        else:
            plot(F[0],F[1],'bo')
    show()
    return nrp
```

```python
print(punctePatrat(500))
```

```python
print(punctePatrat(1000))
```

```python
print(punctePatrat(5000))
```

```python
#b
def punctePatrat2(n):
    axis('square')
    nrp = 0
    axis((0, 1, 0, 1))

    A,B,C,D,E = [0,0],[1,0],[1,1],[0,1],[0.5,0.5]
    for _ in range(n):
        F=[random(),random()]
        if dist(F,A) > dist(F,E) and dist(F,B) > dist(F,E) and dist(F,C) > dist(F,E) and dist(F,D) > dist(F,E) : 
            plot(F[0],F[1],'ro')
            nrp += 1
        else:
            plot(F[0],F[1],'bo')
    show()
    return nrp
```

```python
#b
def punctePatrat3(n):
    axis('square')
    nrp = 0
    axis((0, 1, 0, 1))
    A,B,C,D,E = [0,0],[1,0],[1,1],[0,1],[0.5,0.5]
    for _ in range(n):
        F=[random(),random()]
        FA,FB,FC,FD = dist(F,A), dist(F,B), dist(F,C), dist(F,D)
        res = (FA ** 2 + FB ** 2 < 1) + (FB ** 2 + FC ** 2 < 1) + (FC ** 2 + FD ** 2 < 1) + (FD ** 2 + FA ** 2 < 1)
        if res < 2 : 
            plot(F[0],F[1],'ro')
            nrp += 1
        else:
            plot(F[0],F[1],'bo')
    show()
    return nrp
```
