```python
#1
from scipy.stats import bernoulli, binom
from matplotlib.pyplot import bar, show, hist, grid, legend, xticks
```

```python
def sim_depl(n, p):
    poz = 0
    path = []
    for _ in range(n):
        r = bernoulli.rvs(p)
        if r == 1:
            poz -= 1
        else:
            poz += 1
        path.append(poz)
    #print(path)
    return path
```

```python
#sim_depl(1000, 0.5)
```

```python
def sim_depl_cerc(nr_rep, p, n):
    poz = 0
    path = []
    for _ in range(nr_rep):
        r = bernoulli.rvs(p)
        if r == 1:
            poz -= 1
        else:
            poz += 1
        if poz < 0:
            poz = n + poz
        poz = poz % 360
        path.append(poz)
    #print(path)
    return path
```

```python
#sim_depl_cerc(1000, 0.5, 360)
```

```python
# def sim_depl_cerc(nr_pasi, p, n):
# def sim_depl(nr_pasi, p):
def hist_lin(nr_pasi, p, sim=1000):
    pozitii = [sim_depl(nr_pasi, 0.5)[-1] for _ in range(sim)]
    
    bin_edges = [k+0.5 for k in range(-nr_pasi - 1, nr_pasi + 1)]
    hist(pozitii, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
            alpha = 0.5, label = 'frecvente relative')
    
    distribution = dict([(k - (nr_pasi - k), binom.pmf(k, nr_pasi, sim)) for k in range(nr_pasi + 1)])
    bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
            alpha= 0.6, label = 'probabilitati')
    
    xticks(range(-nr_pasi, nr_pasi +1))
    legend(loc = 'lower left')
    grid()
    show()
```

```python
hist_lin(11, 0.5, sim=1000)
```
